# Sets and Abstract Algebra

## Set Operations in Python

```python
A = {1, 3, 4, 5, 7}
B = {0, 2, 4}
C = set() # empty set
```

A reliable way to remove duplicates from a list is to convert it to a set and back to a list

```python
x = [1, 1, 1, 2, 3, 4, 4, 5, 6, 6]
y = list(set(x))
y
```

The basic set operations include union `|`, intersection `&`, difference `-`, and symmetric difference `^`

```python
A = {1, 3, 4, 5, 7}
B = {0, 2, 4}
```

```python
A | B
```

```python
A & B
```

```python
A - B
```

```python
A ^ B
```

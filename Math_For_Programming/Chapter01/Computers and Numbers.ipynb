# Computers and Numbers

How Python works with binary numbers

```python
converted_binary = 0b10100101
raw_binary = bin(11)
formatted_binary = "{0:b}".format(8675309)


print(f"{converted_binary}\n"
      f"{raw_binary}\n"
      f"{formatted_binary}")
```

How Python works with octal nubmers

```python
converted_octal = 0o1234
raw_octal = oct(77)
formatted_octal = "%o" % 77 # Formats 77 to octal

# When using a leading zero, Python requires a `0x` prefix and `%x` format specifier
converted_octal = 0o01234

print(f"{converted_octal}\n"
      f"{raw_octal}\n"
      f"{formatted_octal}")
```

How Python works with hexadecimal numbers

```python
converted_hex = 0xFDED
converted_hex_2 = 0xdeadbeef
formatted_hex = "%x" % 8675309
raw_hex = hex(8675309)

print(f"{converted_hex}\n"
      f"{converted_hex_2}\n"
      f"{raw_hex}\n"
      f"{formatted_hex}")
```

## Algorithm to convert decimal to binary

```python
def dec2base(d, b):
    def digit(r):
        return chr(48 + r if (r < 10) else 55 + r)

    d, r = (d // b), (d % b)
    m = digit(r)
    while d != 0:
        d, r = (d // b), (d % b)
        m += digit(r)
    return m[::-1]
```

```python
dec2base(6563, 7)
```
